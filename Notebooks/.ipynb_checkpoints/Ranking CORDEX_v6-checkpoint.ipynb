```python
models_dict = {i:name for i,name in enumerate(select_models_folders()[0])}
MODELS_NUMBERS = list(models_dict.keys())
OBS_NUMBER = 17
```

### Loading the data from file

```python
df_all_jja = load_csv2df('models_join_obs_jja.csv', models_output_path, freq_index='3H')
```

```python
df_bias_jja = metric_by_hour(bias, df_all_jja, MODELS_NUMBERS, OBS_NUMBER)
df_sigma_jja = metric_by_hour(sigma_score, df_all_jja, MODELS_NUMBERS, OBS_NUMBER)
df_perkins_jja = metric_by_hour(perkins_skill_score, df_all_jja, MODELS_NUMBERS, OBS_NUMBER)
df_yk_jja = YK_skewness_by_hour(df_all_jja, MODELS_NUMBERS, OBS_NUMBER)
```

## Ranking

### Ranking Bias
* range: -inf +inf
* better: closest to 0
* ascending: False
* strategy: abs.scaling.rounding

```python
bias_rank_jja = set_ranks(df_bias_jja, metric='bias', scale=None, decimals=None)
plot_bump(bias_rank_jja, title="JJA: Bias error ranking",
                    filename='bias_rank_jja', sub_folder='ranking')
```

### Ranking Perkings score
* range: [0,100]
* better: closest to 100
* ascending: True
* strategy: scaling(100).rounding (transforming?)

```python
perkins_rank_jja = set_ranks(df_perkins_jja, metric='perkins', scale=None, decimals=None)
plot_bump(perkins_rank_jja, title="JJA: Perkings score ranking",
                    filename='perkins_rank_jja', sub_folder='ranking')
```

### Ranking Sigma score
* range: -inf +inf
* better: closest to 1
* ascending: False
* strategy: center(0).abs.scaling.rounding

```python
sigma_rank_jja = set_ranks(df_sigma_jja, metric='sigma', scale=None, decimals=None)
plot_bump(sigma_rank_jja, title="JJA: Sigma ratio ranking",
                    filename='sigma_rank_jja', sub_folder='ranking')
```

### Ranking Yule-Kendall skewness score
* range: -inf +inf
* better: closest to 0
* ascending: False
* strategy: abs.scaling.rounding

```python
yk_rank_jja = set_ranks(df_yk_jja, metric='yk', scale=None, decimals=None)
plot_bump(yk_rank_jja, title="Ranking Yule-Kendall skewness score",
                    filename='yk_rank_jja', sub_folder='ranking')
```

### Bump Plots - using plotSlope

```python
# the module plotSlope needs that the index names are of type str
bias_bump_jja = df_bias_jja.copy()
bias_bump_jja.rename(columns = str, index =str, inplace=True)
perkins_bump_jja = df_perkins_jja.copy()
perkins_bump_jja.rename(columns = str, index =str, inplace=True)
sigma_bump_jja = df_sigma_jja.copy()
sigma_bump_jja.rename(columns = str, index =str, inplace=True)
yk_bump_jja = df_yk_jja.copy()
yk_bump_jja.rename(columns = str, index =str, inplace=True)
```
