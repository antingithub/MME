## Averages

```python
# apply abs before avg !!!!!!!!!!!!!!!!


df_cdx_all_met_ensA = [df_cdx_all_met_ensA_djf, df_cdx_all_met_ensA_mam,
df_cdx_all_met_ensA_jja, df_cdx_all_met_ensA_son]

df_sfx_all_met_ensA = [df_sfx_all_met_ensA_djf, df_sfx_all_met_ensA_mam,
df_sfx_all_met_ensA_jja, df_sfx_all_met_ensA_son]

for season,df_met_cdx,df_sfx_met in zip(SEASONS, df_cdx_all_met_ensA, df_sfx_all_met_ensA):
#     df_cdx_all_met_ensA_avg = list(map(lambda df: df.mean(axis=1), df_met_cdx))
#     df_sfx_all_met_ensA_avg = list(map(lambda df: df.mean(axis=1), df_sfx_met))
    df_cdx_all_met_ensA_avg = list(map(lambda df: df.abs().mean(axis=1), df_met_cdx))
    df_sfx_all_met_ensA_avg = list(map(lambda df: df.abs().mean(axis=1), df_sfx_met))
    
    plot_cmp_metric_avg_h(METRIC_NAMES, df_cdx_all_met_ensA_avg, df_sfx_all_met_ensA_avg,
                          title_fig='{}'.format(season.upper()),
                          out_filename='avg_h_cmp_w_ensALPHA_{}'.format(season),
                          sub_folder='metric_ensemble_cmp_avg')
```
