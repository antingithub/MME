#### Each models' mean

```python
df_all_jja.mean(axis=0)
```

```python
df_all_jja_corr = df_all_jja - df_all_jja.mean(axis=0)
df_all_jja_corr.head()
```

```python
save_df2csv(df_all_jja_corr, 'models_join_obs_jja_corr', models_output_path, index=True)
```

```python
df_all_jja_corr_ = load_csv2df('models_join_obs_jja_corr.csv', models_output_path, freq_index='3H')
assert compare_saved_loaded_df(df_all_jja_corr_, df_all_jja_corr)
```

```python
mean_hourly_corr = df_all_jja_corr.groupby(df_all_jja_corr.index.time).mean()
mean_hourly_corr
```

```python
mean_hourly_corr.index =  list(map(lambda t: t.hour, mean_hourly_corr.index))
mean_hourly_corr.index.name = 'hour'

_ = mean_hourly_corr.loc[:, list(range(17))].plot(figsize=(18,8), alpha=0.2)
_ = plt.plot(mean_hourly_corr.mean(axis=1), '-o')
_ = plt.plot(mean_hourly_corr.loc[:, 17], '-o', color='red')
_ = plt.legend(labels=list(mean_hourly_corr.columns[:17])+['models_avg']+['obs'], title="models&obs",
                    loc=1, fontsize='small', fancybox=True)
_ = plt.title("After bias correction: average temperature for 3h time interval. All 16 models and its average." )
_ = plt.xticks(list(range(0,22,3)))
```

* The price to pay for Bias correction is the worst performance during the sun forcing periods (>6h - <18h).
* Is this visible for the corresponding periods for the other seasons? In the winter this should happen for a shorter period.
* Is it better to perform a Bias correction in this season, for 2 periods (>6h - <18h) and (<6h - >18h)?
