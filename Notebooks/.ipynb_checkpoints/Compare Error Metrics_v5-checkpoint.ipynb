## Models' error metrics
* Error metric relative to the OBSERVATIONS

## Bias

#### Bias by hour

```python
MODELS_NUMBERS = list(models_dict.keys())
OBS_NUMBER = 17
MODELS_NUMBERS, OBS_NUMBER
```

```python
df_bias_jja = metric_by_hour(bias, df_all_jja, MODELS_NUMBERS, OBS_NUMBER)
df_bias_jja_CDX = df_bias_jja.abs()
```

### Create artificial random Urb effect for 'JJA'

```python
df_bias_jja_SFX = create_Urb_effect(df_bias_jja).abs()
```

```python
plot_cmp_metric('bias', df_bias_jja_CDX, df_bias_jja_SFX, filename='bias_cmp_jja')
```

### Sigma score - Normalized standard deviation measure

```python
df_sigma_jja = metric_by_hour(sigma_score, df_all_jja, MODELS_NUMBERS, OBS_NUMBER)
df_sigma_jja_CDX = df_sigma_jja.abs()
df_sigma_jja_SFX = create_Urb_effect(df_sigma_jja).abs()
plot_cmp_metric('sigma', df_sigma_jja_CDX, df_sigma_jja_SFX, filename='sigma_cmp_jja')
```

### Perkins skill score

```python
df_perkins_jja = metric_by_hour(perkins_skill_score, df_all_jja, MODELS_NUMBERS, OBS_NUMBER)
df_perkins_jja_CDX = df_perkins_jja.abs()
df_perkins_jja_SFX = create_Urb_effect(df_perkins_jja).abs()
plot_cmp_metric('perkins', df_perkins_jja_CDX, df_perkins_jja_SFX, filename='perkins_cmp_jja')
```

### Yule-Kendall skewness 

```python
df_yk_jja = YK_skewness_by_hour(df_all_jja, MODELS_NUMBERS, OBS_NUMBER)
df_yk_jja_CDX = df_yk_jja.abs()
df_yk_jja_SFX = create_Urb_effect(df_yk_jja).abs()
plot_cmp_metric('yk', df_yk_jja_CDX, df_yk_jja_SFX, filename='yk_cmp_jja')
```
