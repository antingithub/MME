```python
def select_models_folders(dir_models = '..\data\models', key_word = 'historical'):
    folders = [folder for folder in  os.listdir(os.path.join(os.getcwd(), dir_models)) if key_word in folder]
    #print('Total number of models: {}'.format(len(folders)))
    #print(folders)
    return folders, len(folders)

folders, nmodels = select_models_folders('..\data\models', 'historical')
folders, nmodels
```

```python
def make_dates4models(ndata, year_ini = 1971, hours_interval = 3):
    '''
        ndata: number of data points
        year_ini = 1971
        hours_interval = 3
    '''
    month, day, hour = 1, 1, hours_interval          # first hour is 3 (data is every 3 hours)
    DataV=[]
    for t in range(0, ndata):
        DataV.append([year_ini,month,day,hour])
        hour  += 3
        if hour > 21:
            hour = 0
            day += 1
            if day > calendar.monthrange(year_ini, month)[1]:
                day = 1
                month += 1
                if month > 12:
                    month = 1
                    year_ini += 1
    DataV = [[1971,    1,    1,    0]] + DataV
    return np.array(DataV)

#assert make_dates4models(102271).shape == (102272, 4)
# d = make_dates4models(102271)
# d[:10], d[-10:]
# array([[1971,    1,    1,    0],
#         [1971,    1,    1,    3],
#         [1971,    1,    1,    6],
#         [1971,    1,    1,    9],
#         [1971,    1,    1,   12],
#         [1971,    1,    1,   15],
#         [1971,    1,    1,   18],
#         [1971,    1,    1,   21],
#         [1971,    1,    2,    0],
#         [1971,    1,    2,    3]
#       ...
#         [2005,   12,   30,   18],
#         [2005,   12,   30,   21],
#         [2005,   12,   31,    0],
#         [2005,   12,   31,    3],
#         [2005,   12,   31,    6],
#         [2005,   12,   31,    9],
#         [2005,   12,   31,   12],
#         [2005,   12,   31,   15],
#         [2005,   12,   31,   18],
#         [2005,   12,   31,   21]])
```

```python
def get_data_models(dir_models = '..\data\models', var_filename = 'Forc2m_TA.txt', start_year = 1979):
    '''
        For temperature only!! Convertion to C
        start_year = 1979
        end_year = 2005
        
        Comments:
        OBS1 - Why, T2m = np.append(T2m,np.nan)?
        Because the dataset has no value for the last point for 21-0h
        [2006    1    1    0]
        
    '''
    # parameters
    hours = np.arange(0,22,3)
    
    
    # get the models' folders
    folders, nmodels = select_models_folders(dir_models, 'historical')
    
    # from the first folder in folders the var_filename i sused to get the number of data points
    data = np.loadtxt(os.path.join(dir_models, folders[0], var_filename))
    Lt = len(data)
    
    DateV = make_dates4models(Lt, 1971)
    # build a df with ('year', 'month', 'day') only. Assume for every day there the hour = 0 exists always!
    DateV = DateV[DateV[:,3] == 0][:,:3]
    df_date = pd.DataFrame(DateV, columns = ['year', 'month', 'day'], dtype=int)
    # add seasons column
    conditions = [(df_date['month'].isin([12, 1, 2])), (df_date['month'].isin([3, 4, 5])),
                (df_date['month'].isin([6, 7, 8])), (df_date['month'].isin([9, 10, 11]))]
    df_date['season'] = np.select(conditions, ['DJF','MAM','JJA','SON'])
    
    models = []
    for cmod,folder in enumerate(folders):   # REMOVE
        T2m = np.loadtxt(os.path.join(dir_models, folder, var_filename)) - 273.15  # and convert temperature to C
        T2m = np.append(np.asarray([np.nan]), np.asarray(T2m))    # See OBS1 in comments
        df_model = pd.DataFrame(T2m.reshape(-1,8), columns = hours)
        df_model = pd.concat([df_date, df_model], axis=1)
        models.append(df_model)
            
    return models    
```

```python
def compute_pdf_cdf(sample, plot=True):
    '''
        Visualize the ECDF, interpolated PDF and samples
        
    '''
    bins=int(np.sqrt(len(sample)))
#     bins =  100
    hist = np.histogram(sample, bins = bins)
    hist_dist = st.rv_histogram(hist)

    sample_min, sample_max = np.min(sample), np.max(sample)
    pdf, Ecdf = hist_dist.pdf, hist_dist.cdf
    assert Ecdf(sample_min) == 0, 'Something wrong with ECDF!'
    assert Ecdf(sample_max) == 1, 'Something wrong with ECDF!'

    if plot:
        X = np.linspace(sample_min, sample_max, bins)
        plt.title("PDF")
        _ = plt.hist(sample, density = True, bins = bins)
        _ = plt.plot(X, pdf(X), label = 'PDF')
        #_ = plt.plot(X, Ecdf(X), label = 'CDF')
        plt.legend()
        plt.margins(0.02)
        plt.show()
    return pdf, Ecdf
```

### For each model, all hours (pooling together)

```python
for cmod,model in enumerate(all_models_jja):
    print('PDF for model {}'.format(cmod))
    all_hours = model.iloc[:,4:].values.reshape(-1)
    _, _ = compute_pdf_cdf(all_hours, plot=True)
```

### Same as before but using KDE (ML)
*For each model, all hours (pooling together)

```python
def compute_pdf_KDE(sample, bandwidth=2, kernel='gaussian', plot=True):
    '''
    
    '''
    bins=int(np.sqrt(len(sample)))
    #print(bins)
    sample = sample.reshape((len(sample), 1))
    # fit density
    model = KernelDensity(bandwidth=bandwidth, kernel=kernel)
    model.fit(sample)
    
    if plot:
        sample_min, sample_max = np.min(sample), np.max(sample)
        X = np.linspace(sample_min, sample_max, bins)
        values = X
        values = values.reshape((len(values), 1))
        #print(values.shape)
        probabilities = model.score_samples(values)
        probabilities = np.exp(probabilities)
        _ = plt.hist(sample, bins=bins, density=True)
        _ = plt.plot(values[:], probabilities)
        plt.title("pdf using KDE")
        plt.show()
    #return np.exp(model.score_samples)
    return model.score_samples
```

```python
for cmod,model in enumerate(all_models_jja):
    print('PDF for model {}'.format(cmod))
    all_hours = model.iloc[:,4:].values.reshape(-1)
    _ = compute_pdf_KDE(all_hours, plot=True)
```

### For each model pooling together by hour

```python
for cmod,model in enumerate(all_models_jja):
    all_hours = model.iloc[:,4:]
    for hour in all_hours:
        print('PDF for model = {} @ hour = {}'.format(cmod, hour))
        one_hour = all_hours[hour].values
        _ = compute_pdf_KDE(one_hour, plot=True)
```
