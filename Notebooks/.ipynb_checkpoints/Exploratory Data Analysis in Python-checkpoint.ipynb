## Resampling
* Some groups are oversampling 

```python
def resample_rows_weighted(df, column='finalwgt', seed=17):
    """Resamples a DataFrame using probabilities proportional to given column.
        df: DataFrame
        column: string column name to use as weights
        returns: DataFrame
    """
    np.random.seed(seed)
    weights = df[column] / sum(df[column])                                               # TRICK - the weights are the fractions
    
    indices = np.random.choice(df.index, len(df), replace=True, p=weights)            # TRICK - resampling indexes w rep and prob
    sample = df.loc[indices]                                                             # TRICK
    return sample
```

```python
# the dataset is not representative of the population, so we have to correct
# 'wgt2013_2015' - sampling weights

sample = resample_rows_weighted(nsfg, 'wgt2013_2015', seed=17)
sample.head()
```

```python
sample.shape, sum(sample.duplicated())
```

```python
# confirm that it made sampling with replacement!
sample.index.value_counts().sort_values(ascending=False).head()
```

```python
pounds  = sample['birthwgt_lb1'].replace([98,99], np.nan)
ounces  = sample['birthwgt_oz1'].replace([98,99], np.nan)
birth_weight = pounds + ounces/16
```

```python
# Filter preterm babies
preterm = sample['prglngth'] < 37

# Filter single births
single = sample['nbrnaliv'] == 1

# Compute birth weight for single full-term babies
single_full_term_weight = birth_weight[single&~preterm]
print('Single full-term mean:', single_full_term_weight.mean())

# Compute birth weight for multiple full-term babies
mult_full_term_weight = birth_weight[~single&~preterm]
print('Multiple full-term mean:', mult_full_term_weight.mean())
```

```python
def my_pmf(s, normalize=False):
    c = s.value_counts(dropna=True)
    if normalize:
        c = c/c.sum()
    return c.sort_values(ascending=True)
    
```

```python
pmf_educ = my_pmf(educ)
```

```python
pmf_educ[12]
```

```python
pmf_educ= my_pmf(educ, 1)
pmf_educ[12]
```

```python
# the same as above but with my_pmf()

plt.bar(pmf_educ.index, pmf_educ);
plt.xlabel('Years of education')
plt.ylabel('PMF')
plt.show()
```

```python
age = gss['age']

pmf_age = my_pmf(age)

plt.bar(pmf_age.index, pmf_age)
plt.xlabel('Age')
plt.ylabel('PMF')
plt.show()
```

### CDF - turns any SEQUENCE in a cummulative prob

```python
def my_cdf(s):
    from operator import add
    from itertools import accumulate
    pmf = my_pmf(s, normalize=True).sort_index(ascending=True)
    acc = accumulate(pmf, add)
    return pd.Series([*acc], index=pmf.index)
```

```python
cdf_age = my_cdf(age)
cdf_age.iloc[:10]
```

#### Holding educ constant

```python
educ_years = [12, 14, 16]

pred_df = pd.DataFrame()
pred_df['age'] = np.linspace(20, 90, 71, dtype=int)
pred_df['age2'] = pred_df['age']**2

plt.plot(gss.groupby('age').realinc.median(), 'o', alpha=0.3)
plt.plot(gss.groupby('age').realinc.mean(), 'o', alpha=0.3)
for edu_y in educ_years:
    pred_df['educ'] = edu_y
    pred_df['educ2'] = edu_y**2
    pred_df['realinc_pred'] = results3.predict(pred_df)
    plt.plot('age', 'realinc_pred', data = pred_df )
    
plt.ylabel('Real income')
plt.xlabel('Age')
plt.legend(['median', 'mean', '12_years', '14_years', '16_years'])   
plt.show()

```
